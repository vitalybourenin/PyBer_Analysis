### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
urban_cities_df = pyber_data_df[pyber_data_df["type"]=="Urban"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"]=="Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"]=="Rural"]
```

```python
# 2. Get the total drivers for each city type
urban_driver_count= urban_cities_df.groupby(["city"]).mean()["driver_count"].sum()
suburban_driver_count= suburban_cities_df.groupby(["city"]).mean()["driver_count"].sum()
rural_driver_count = rural_cities_df.groupby(["city"]).mean()["driver_count"].sum()
```

```python
total_drivers_by_type = city_data_df.groupby(["type"]).sum()['driver_count']
total_drivers_by_type
```

```python
#  3. Get the total amount of fares for each city type
urban_fare_total = urban_cities_df.sum()["fare"]
suburban_fare_total = suburban_cities_df.sum()["fare"]
rural_fare_total = rural_cities_df.sum()["fare"]
```

```python
total_fare_by_type = pyber_data_df.groupby(["type"]).sum()['fare']
total_fare_by_type
```

```python
#  4. Get the average fare per ride for each city type. 
#Find total number of rides by city type
urban_total_rides = urban_cities_df.count()["ride_id"].sum()
suburban_total_rides = suburban_cities_df.count()["ride_id"].sum()
rural_total_rides= rural_cities_df.count()["ride_id"].sum()
```

```python
total_rides_by_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_by_type
```

```python
urban_avg_fare = urban_fare_total / urban_total_rides
suburban_avg_fare = suburban_fare_total / suburban_total_rides
rural_avg_fare = rural_fare_total / rural_total_rides
```

```python
avg_fare_by_type = pyber_data_df.groupby(["type"]).mean()["fare"]
avg_fare_by_type
```

```python
# 5. Get the average fare per driver for each city type. 
urban_avg_fpd = urban_fare_total / urban_driver_count
suburban_avg_fpd = suburban_fare_total / suburban_driver_count
rural_avg_fpd = rural_fare_total / rural_driver_count
```

```python
avg_fpd_by_type = total_fare_by_type / total_drivers_by_type
avg_fpd_by_type
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame(
    {"Total Rides": total_rides_by_type,
    "Total Drivers": total_drivers_by_type,
    "Total Fares": total_fare_by_type,
    "Average Fare per Ride": avg_fare_by_type,
    "Average Fare per Driver": avg_fpd_by_type, }
)
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
format_mapping={'Total Fares': '${:,.2f}', 'Average Fare per Ride': '${:,.2f}', 'Average Fare per Driver': '${:,.2f}'}

for key, value in format_mapping.items():
    pyber_summary_df[key] = pyber_summary_df[key].apply(value.format)
    
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
new_pyber_data_df = pyber_data_df.groupby(['type', 'date']).sum()['fare']
new_pyber_data_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
new_pyber_data_df = new_pyber_data_df.reset_index()
new_pyber_data_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
p_table = pd.pivot_table(new_pyber_data_df, index=['date'], columns=['type'], values=['fare'])
p_table
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
date_range_df = p_table.loc['2019-01-01' : '2019:04-29']
date_range_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
date_range_df.index = pd.to_datetime(date_range_df.index)
date_range_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
date_range_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
sum_fares_weekly_df = date_range_df.resample("W").sum()
sum_fares_weekly_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

ax = sum_fares_weekly_df.plot(figsize=(18, 8))

ax.set_xlabel('Date')
ax.set_ylabel('Fares ($USD)')
ax.set_title("Total Fare by City Type")
ax.legend()

plt.savefig("analysis/weekly_fare_summary.png")

plt.show()
```
